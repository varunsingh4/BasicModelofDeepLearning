Linear Regression with Python Scikit Learn
In this section we will see how the Python Scikit-Learn library for machine learning can be used to implement regression functions. We will start with simple linear regression involving two variables.


# Data Preprocessing

```python
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python
train_samples=[]
test_samples=[]
train_labels=[]
test_labels=[]
```

Example of the data
- An experimental drug was tested on individuals from ages 13 to 100 in a clinical trial.
- The trial had 2100 participants.Half were under 65 years old,half were 65 years or older.
- Around 95% of patients 65 or older experienced side effects
- Around 95% of patients under 65 experienced no effects

```python
for i in range(50):
    Younger_individuals=randint(13,64)      #5% of younger individuals who did experienced side effects
    train_samples.append(Younger_individuals)
    train_labels.append(1)
    #5% of older individuals who did not experience side effects
    Older_individuals=randint(65,100)
    train_samples.append(Older_individuals)
    train_labels.append(0)
    
for i in range(1000):
    #95% of younger individuals who did not experience side effects
    random_younger=randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)
    #95% of older individuals who did experience side effects
    random_older=randint(65,100)
    train_samples.append(random_older)
    train_labels.append(1)
    
    
```

```python
for i in train_samples:
      print(i)
```

```python
for i in train_labels:
    print(i)
```

```python
train_labels=np.array(train_labels)
train_samples=np.array(train_samples)
train_labels,train_samples=shuffle(train_labels,train_samples)
```

```python
scaler=MinMaxScaler(feature_range=(0,1))
scaler_train_samples=scaler.fit_transform(train_samples.reshape(-1,1))
```

```python
for i in scaler_train_samples:
    print(i)
```

# SIMPLE KERAS MODEL

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation,Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
physical_devices=tf.config.experimental.list_physical_devices('GPU')
print('Number of GPU available',len(physical_devices))
tf.config.experimental.set_memory_growth(physical_devices[0],True)
```

```python
model= Sequential([
    Dense(units=16, input_shape=(1,),activation='relu'),
    Dense(units=32,activation='relu'),
    Dense(units=2,activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.001),loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x=scaler_train_samples, y=train_labels,validation_split=0.1,batch_size=10,epochs=30,shuffle=True,verbose=True)
```
